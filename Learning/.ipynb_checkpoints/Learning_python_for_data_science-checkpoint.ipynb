The Structured Query Language (SQL) lets you perform all sorts of management tasks in a relational database, retrieve data as needed, and even shape it in a particular way so that the need to perform additional shaping is unnecessary

Creating a connection to a database can be a complex undertaking. For
one thing, you need to know how to connect to that particular database.
However, you can divide the process into smaller pieces. The first step is
to gain access to the database engine. You use two lines of code similar to
the following code (but the code presented here is not meant to execute
and perform a task):

from sqlalchemy import create_engine
engine = create_engine('sqlite:///:memory:')

After you have access to an engine, you can use the engine to perform
tasks specific to that DBMS. The output of a read method is always a
DataFrame object that contains the requested data. To write data, you must
create a DataFrame object or use an existing DataFrame object. You
normally use these methods to perform most tasks:

read_sql_table(): Reads data from a SQL table to a DataFrame
object
read_sql_query(): Reads data from a database using a SQL query to
a DataFrame object

read_sql(): Reads data from either a SQL table or query to a
DataFrame object

DataFrame.to_sql(): Writes the content of a DataFrame object to the
specified tables in the database

The sqlalchemy library provides support for a broad range of SQL
databases. The following list contains just a few of them:

SQLite
MySQL
PostgreSQL
SQL Server

Other relational databases, such as those you can connect to using
Open Database Connectivity (ODBC)

XML is the main form that web data comes in

```python
from lxml import objectify
import pandas as pd

xml = objectify.parse(open('XMLData.xml'))
root = xml.getroot() #determines the "root" of the data tree
df = pd.DataFrame(columns=('Number', 'String', 'Boolean')) #tells the columns of data

for i in range(0,4):
    obj = root.getchildren()[i].getchildren()
    row = dict(zip(['Number', 'String', 'Boolean'],
    [obj[0].text, obj[1].text,
    obj[2].text]))
    row_s = pd.Series(row)
    row_s.name = i
    df = pd.concat([df, pd.DataFrame([row_s])], ignore_index=True)
    
print(df)
```

The following code is meant to help with deleting repeat data from an XML data source to help clean it:

```python
from lxml import objectify
import pandas as pd

xml = objectify.parse(open('XMLData2.xml'))
root = xml.getroot()
df = pd.DataFrame(columns=('Number', 'String', 'Boolean'))

for i in range(0,4):
    obj = root.getchildren()[i].getchildren()
    row = dict(zip(['Number', 'String', 'Boolean'],
    [obj[0].text, obj[1].text,
    obj[2].text]))
    row_s = pd.Series(row)
    row_s.name = i
    df = pd.concat([df, pd.DataFrame([row_s])], ignore_index=True)
    
search = pd.DataFrame.duplicated(df)
print(df)
print()
print(search[search == True])
```

Remove the duplicate row from the XML file:

```python
from lxml import objectify
import pandas as pd

xml = objectify.parse(open('XMLData2.xml'))
root = xml.getroot()
df = pd.DataFrame(columns=('Number', 'String', 'Boolean'))

for i in range(0,4):
    obj = root.getchildren()[i].getchildren()
    row = dict(zip(['Number', 'String', 'Boolean'],
    [obj[0].text, obj[1].text,
    obj[2].text]))
    row_s = pd.Series(row)
    row_s.name = i
    df = pd.concat([df, pd.DataFrame([row_s])], ignore_index=True)
print(df.drop_duplicates())
```

The following code is for using the Bag of Words model for natural language processing:

```python
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import *

categories = ['comp.graphics', 'misc.forsale',
    'rec.autos', 'sci.space']
twenty_train = fetch_20newsgroups(subset='train',
    categories=categories,
    shuffle=True,
    random_state=42)
count_vect = CountVectorizer() #initializes the variable as a CountVectorizer object
X_train_counts = count_vect.fit_transform( #places the train data into a variables using a method on the object
    twenty_train.data)

print("BOW shape:", X_train_counts.shape) #gives the matrix dimensions?
caltech_idx = count_vect.vocabulary_['caltech'] #assigns the data associated with 'caltch' to a variable
print('"Caltech": %i' % X_train_counts[0, caltech_idx]) #gives the numer of times a words has appeared in the data set
```

n-grams are used for predicting the next input from a user based on previous patterns. This is useful for search engine predictions.

```python
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import * #fetch the data set

categories = ['sci.space']

twenty_train = fetch_20newsgroups(subset='train',
    categories=categories,
    remove=('headers',
    'footers',
    'quotes'),
    shuffle=True,
    random_state=42) #removes words from the imported data set

count_chars = CountVectorizer(analyzer='char_wb', #parameter looks at the character count within word boundaries
    ngram_range=(3,3), #parameter determines the min,max size of an n-gram; gives character tri-grams
    max_features=10) #determines how many features the vectorizer returns; gives the top ten in this case

count_chars.fit(twenty_train['data'])

count_words = CountVectorizer(analyzer='word',
    ngram_range=(2,2), #gives n-grams for the most frequent words, rather than characters appearing together like above
    max_features=10,
    stop_words='english') #removes words from the data set that are part of the english "pickle"
count_words.fit(twenty_train['data'])
X = count_chars.transform(twenty_train.data)
print(count_chars.get_feature_names_out())
print(X[1].todense()) #includes data that is missing in the set without values
print(count_words.get_feature_names_out())
```

Term Frequency times Inverse Document Frequency (TF-IDF) helps to give words that are most important in the document. This helps to
reduce a document to the least amount of information to help give meaning with less "fluff" information (stop words). VERY helpful for search
engines looking for documents that fit the search parameters

```python
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import *

categories = ['comp.graphics', 'misc.forsale',
    'rec.autos', 'sci.space'] #gives the categories as before
twenty_train = fetch_20newsgroups(subset='train',
    categories=categories,
    shuffle=True,
    random_state=42)

count_vect = CountVectorizer()
X_train_counts = count_vect.fit_transform(
    twenty_train.data) #places the data into CountVectorizer object

tfidf = TfidfTransformer().fit(X_train_counts) #Term Frequency times Inverse Document Frequency (TF-IDF)
X_train_tfidf = tfidf.transform(X_train_counts)

caltech_idx = count_vect.vocabulary_['caltech']
print('"Caltech" scored in a BOW:')
print('count: %0.3f' % X_train_counts[0, caltech_idx])
print('TF-IDF: %0.3f' % X_train_tfidf[0, caltech_idx])
```

"term frequencyinverse document frequency, is a measure of importance of a word to a document in a collection or corpus, adjusted for the fact that some words appear more frequently in general"

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

spread = 100 * np.random.rand(100) #gives 100 random values
center = np.ones(50) * 50 #produces 50 values in the center of the range of the set
flier_high = 100 * np.random.rand(10) + 100 #simulates high outliers
flier_low = -100 * np.random.rand(10) #simulates low outliers

data = np.concatenate((spread, center,
    flier_high, flier_low)) #combines all the data in one set

plt.boxplot(data, sym='gx', widths=.75, notch=True) #produces a box plot
plt.show()
```

This code takes the data and creates a trend line to represent it:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
%matplotlib inline

start_date = dt.datetime(2018, 7, 29)
end_date = dt.datetime(2018, 8, 7)
daterange = pd.date_range(start_date, end_date)
sales = (np.random.rand(len(daterange)) * 50).astype(int)
df = pd.DataFrame(sales, index=daterange,
    columns=['Sales'])

lr_coef = np.polyfit(range(0, len(df)), df['Sales'], 1)
lr_func = np.poly1d(lr_coef)
trend = lr_func(range(0, len(df)))
df['trend'] = trend

df.loc['Jul 30 2018':'Aug 05 2018'].plot()
plt.xlabel('Sales Date')
plt.ylabel('Sale Value')
plt.title('Plotting Time')
plt.legend(['Sales', 'Trend'])
plt.show()
```
